```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#df_Bitcoins = pd.read_csv('BTC_EUR.csv')
df_Bitcoins = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
```

```python
df_Bitcoins.info()
```

```python
plt.figure(figsize=(10,8))
df_Bitcoins['2022']['Close'].plot()
```

```python
df_Bitcoins.loc['2023', 'Close'].resample('M').plot()
plt.show()
```

```python
df_Bitcoins.loc['2023', 'Close'].resample('W').mean().plot()
plt.show()
```

```python
df_Bitcoins.loc['2022', 'Close'].resample('W').agg(['mean','min','max']).plot()
```

```python
df_Etherium = pd.read_csv('ETH-EUR.csv', index_col='Date', parse_dates=True)
```

```python
df_eth_btc = pd.merge(df_Bitcoins, df_Etherium, on='Date',how='inner', suffixes=('_btc', '_eth'))
```

```python
df_eth_btc
```

```python
df_eth_btc[['Close_btc','Close_eth']].plot(subplots=True, figsize=(10,8))
```

```python
df_eth_btc[['Close_btc','Close_eth']].corr()
```
